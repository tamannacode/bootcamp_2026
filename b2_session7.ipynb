```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Redefine the 'students' DataFrame with numerical data from PzBCoyeArT5h to ensure its state
students = pd.DataFrame({
    "name": ["Aarav", "Riya", "Kabir", "Meera", "Dev", "Sara"],
    "study_hours": [6, 3, 8, 2, 5, 7],
    "attendance": [85, 70, 90, 60, 78, 92],
    "previous_marks": [78, 55, 88, 35, 65, 84],
    "final_marks": [82, 58, 91, 40, 70, 88]
})

feature_columns=["study_hours","attendance","previous_marks"]
students_vectors=students[feature_columns].values


# Create a new DataFrame for skills
skill_students = pd.DataFrame({
    "name": ["Aarav", "Riya", "Kabir", "Meera", "Dev", "Sara"],
    "skills": [
        "Python Machine Learning AI",
        "Python AI Data Science",
        "Machine Learning Python",
        "Java Web Development",
        "Python AI ML",
        "Python Machine Learning AI"
    ]
})

vectorizer = TfidfVectorizer()
student_vectors_skills = vectorizer.fit_transform(skill_students["skills"])

similarity_matrix = cosine_similarity(student_vectors_skills)

similarity_df = pd.DataFrame(
    similarity_matrix,
    index=skill_students["name"],
    columns=skill_students["name"]
)

print("Numerical Student Vectors:\n", students_vectors)
print("\nSkill Similarity Matrix:\n", similarity_df)
```

```python
vocab = {
    "the":1,
    "movie":2,
    "was":3,
    "not":4,
    "good":5
}

# Preprocess tokens to handle 'n=movies' explicitly
input_tokens = ["the", "n-movies", "was", "not", "good"]
processed_tokens = []
for word in input_tokens:
    if word == "n-movies":
        processed_tokens.append("movie") # Replace 'n=movies' with 'movie'
    elif word in vocab:
        processed_tokens.append(word)
    # Optional: handle words not in vocab, e.g., skip them or assign an 'UNKNOWN' token ID

token_ids =[vocab[word] for word in processed_tokens]
print(token_ids)
# Expected output: [1,2,3,4,5]
```

```python
positive_words=["good","great","amazing","excellent","love"]
negative_words=["bad","terrible","awful","poor","hate","boring"]


def simple_sentimental(sentence):
  words = sentence.lower().replace(".","").split()

  score=0

  for word in words:
    if word in positive_words:
      score+=1
    elif word in negative_words:
      score-=1
  if score>0:
    return "Positive"
  elif score<0:
    return "Negative"
  else:
    return "Natural"

print(simple_sentimental("The movie was good"))
print(simple_sentimental("The movie was bad"))
print(simple_sentimental("The movie was not good"))
```

```python
def better_sentimental_with_not(sentence):
  words = sentence.lower().replace(".","").split()
  score=0



  for i, word in enumerate(words):
    if word in positive_words:
      if i>0 and words[i-1]=="not":
        score-=1
      else:
        score+=1
    elif word in negative_words:
      if i>0 and words[i-1]=="not":
        score+=1
      else:
        score-=1
  if score>0:
    return "Positive"
  elif score<0:
    return "Negative"
  else:
    return "Natural"

print(better_sentimental_with_not("The movie was good"))
print(better_sentimental_with_not("The movie was not good"))
print(better_sentimental_with_not("The movie was not not good"))
```
